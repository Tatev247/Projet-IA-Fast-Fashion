## 4. Analyse et Visualisation des Données.

### A. Analyse univariée.

##### Prix_en_euros :

```python
# Résumé statistique de prix_en_euros :
zyga['prix_en_euros'].describe()
```

On peut dire que la colonne "prix_en_euros" contient 108 observations non manquantes, avec une moyenne de 159,52 euros et un écart type de 54,70 euros. La plus petite valeur est 54 euros et la plus grande est 288 euros. La moitié des valeurs se situe entre 138 et 168 euros, et 25% des valeurs sont inférieures à 138 euros tandis que 25% sont supérieures à 178 euros.

```python
# Calcul des coefficients de skewness et kurtosis
skewness = zyga['prix_en_euros'].skew()
kurtosis = zyga['prix_en_euros'].kurtosis()
print('Skewness:', skewness)
print('Kurtosis:', kurtosis)
```

```python
# Boxplot de prix_en_euros :
plt.boxplot(zyga['prix_en_euros'])
plt.title('Distribution des prix')
plt.show()
```

```python
# Visualisation des prix :
plt.hist(zyga['prix_en_euros'], bins=20)
plt.xlabel('Prix')
plt.ylabel('Nombre d\'articles')
plt.title('Distribution des prix')
plt.show()
```

```python
sns.kdeplot(data=zyga, x='prix_en_euros')
plt.xlabel('Prix en euros')
plt.title('Distribution des prix en euros')
plt.show()
```

En combinant les résultats de la fonction describe(), les coefficients de skewness et kurtosis et les visualisations, on peut conclure que la distribution des prix_en_euros est légèrement asymétrique à gauche et est relativement aplatie. Cela signifie que la majorité des prix sont situés autour de la moyenne et que la distribution a relativement peu de données aberrantes. L'intervalle interquartile est relativement étroit, ce qui confirme que la majorité des prix se situent dans un intervalle relativement restreint. Toutefois, il y a quelques valeurs extrêmes (outliers) dans la distribution, ce qui affecte la moyenne. Globalement, la distribution semble être assez proche d'une distribution normale, bien que ce ne soit pas parfait.

```python
# prix_en_euros suit une loi normale ? : test de Shapiro-Wilk

from scipy.stats import shapiro

# Test de Shapiro-Wilk pour la normalité
stat, p = shapiro(zyga['prix_en_euros'])
print('Statistiques du test de Shapiro-Wilk:', stat)
print('p-value:', p)
if p > 0.05:
    print('La variable suit une distribution normale')
else:
    print('La variable ne suit pas une distribution normale')
```

```python
# Calculer l'intervalle de confiance à 90% de la moyenne de prix_en_euros :
import scipy.stats as stats

n = len(zyga['prix_en_euros'])
mean = zyga['prix_en_euros'].mean()
std_error = zyga['prix_en_euros'].std() / (n**0.5)
confidence_interval = stats.t.interval(0.9, n-1, loc=mean, scale=std_error)
print('Intervalle de confiance à 90%:', confidence_interval)
```

### B. Analyse bivariée.

#### Prix_en_euros & Composition :

```python
# Création d'un DataFrame à partir de la moyenne de prix par composition et tri par ordre croissant de la moyenne de prix :
zyga_mean_price_by_comp_sorted = pd.DataFrame(mean_price_by_comp).sort_values(by="prix_en_euros")
zyga_mean_price_by_comp_sorted
```

```python
# Affichage des moyennes de prix par composition triées dans un bar plot :
plt.bar(zyga_mean_price_by_comp_sorted.index, zyga_mean_price_by_comp_sorted["prix_en_euros"])
plt.xticks(rotation=90)
plt.title("Moyenne de prix par composition")
plt.xlabel("Composition")
plt.ylabel("Prix en euros")
plt.show()
```

```python
# Création d'un box plot pour visualiser la distribution des prix par composition :
sns.boxplot(x="composition", y="prix_en_euros", data=zyga_comp_prix)
plt.xticks(rotation=90)
plt.title("Distribution des prix par composition")
plt.xlabel("Composition")
plt.ylabel("Prix en euros")
plt.show()
```

```python
# Création d'un violin plot pour visualiser la densité de probabilité des prix par composition :
sns.violinplot(x="composition", y="prix_en_euros", data=zyga_comp_prix)
plt.xticks(rotation=90)
plt.title("Densité de probabilité des prix par composition")
plt.xlabel("Composition")
plt.ylabel("Prix en euros")
plt.show()
```

D'après les informations fournies par la réponse, on peut constater que les articles ayant une composition de "100% Lin Léger (135g/m²)" ont le prix moyen le plus élevé (191,28 euros), tandis que ceux ayant une composition de "100% Jersey de Lin" ont le prix moyen le plus bas (92,75 euros). Les articles en "100% Gaze de Lin (85g/m2) imprimée et lavée" ont un prix moyen plus élevé que ceux en "100% Gaze de Lin (85g/m2)". Les articles en "100% Lin Moyen (180g/m²)" ont un prix moyen de 167,31 euros.
