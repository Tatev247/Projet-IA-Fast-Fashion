### A. Analyse univariée.

##### Product_price :

```python
# Résumé statistique de product_price :
hm['product_price'].describe()
```

```python
# Filtrer les lignes avec des valeurs aberrantes de prix
seuil_sup = 100
lignes_aberrantes = (hm['product_price'] < 0) | (hm['product_price'] > seuil_sup)

# Supprimer les lignes aberrantes
hm = hm.drop(hm[lignes_aberrantes].index)
hm
```

```python
# Résumé statistique de product_price :
hm['product_price'].describe()
```

On remarque qu'il y a 203 valeurs non manquantes pour les prix des produits dans l'échantillon. La moyenne des prix est d'environ 32.72, avec un écart type de 11.13, ce qui indique une certaine variation des prix. Le prix le plus bas observé est de 5.99, tandis que le prix le plus élevé est de 49.99. Environ 25% des prix sont inférieurs à 24.99, 50% sont inférieurs à 34.99 et 75% sont inférieurs à 39.99.

```python
# Calcul des coefficients de skewness et kurtosis
skewness = hm['product_price'].skew()
kurtosis = hm['product_price'].kurtosis()
print('Skewness:', skewness)
print('Kurtosis:', kurtosis)
```

La distribution des prix semble légèrement asymétrique avec une légère queue vers la gauche, comme indiqué par un coefficient de skewness de -0.48. Cela suggère une légère prédominance de prix inférieurs à la moyenne.
Le coefficient de kurtosis de -0.85 indique une distribution légèrement moins étalée (aplatissement) par rapport à une distribution normale.

```python
# Boxplot de prix_en_euros :
plt.boxplot(hm['product_price'])
plt.title('Distribution des prix')
plt.show()
```

```python
# Visualisation des prix :
plt.hist(hm['product_price'], bins=20)
plt.title('Distribution des prix des produits')
plt.xlabel('Prix ($)')
plt.ylabel('Nombre de produits')
plt.show()
```

```python
# Tracer l'estimation de la densité de la colonne "product_dollar_price"
sns.kdeplot(hm['product_dollar_price'], shade=True)

# Ajouter un titre et des labels pour les axes
plt.title("Estimation de densité de product_dollar_price")
plt.xlabel("Prix en dollars")
plt.ylabel("Densité estimée")
```

En combinant les résultats de la fonction describe(), les coefficients de skewness et kurtosis et les visualisations, on peut conclure que les données montrent que la plupart des prix des produits se situent autour de 32.72, avec une légère dispersion. Le prix le plus bas observé est de 5.99 et le prix le plus élevé est de 49.99. La distribution des prix semble légèrement asymétrique, avec une prédominance de prix inférieurs à la moyenne. Globalement, les prix sont relativement proches de la moyenne, avec quelques prix extrêmes.

```python
# Test shapiro :
from scipy.stats import shapiro

# Effectuer le test de Shapiro-Wilk sur la distribution des prix en dollars
stat, p = shapiro(hm['product_price'])

# Interpréter le résultat du test
alpha = 0.05
if p > alpha:
    print('La distribution des prix en dollars semble être normale (p-value = {:.4f})'.format(p))
else:
    print('La distribution des prix en dollars ne semble pas être normale (p-value = {:.4f})'.format(p))
```

#### Product_category & Composition :

```python
import matplotlib.pyplot as plt

# Sélectionner les colonnes product_category et composition
categories = hm[['product_name', 'cotton']]

# Compter le nombre d'occurrences de chaque catégorie
category_counts = categories['product_name'].value_counts()

# Créer un dictionnaire pour stocker les pourcentages de composition par catégorie
category_compositions = {}

# Pour chaque catégorie, calculer le pourcentage de chaque composition
for category in category_counts.index:
    compositions = categories[categories['product_name'] == category]['cotton']
    composition_counts = compositions.value_counts(normalize=True)
    category_compositions[category] = composition_counts

# Afficher un graphique à barres pour chaque catégorie avec ses pourcentages de composition correspondants
fig, axes = plt.subplots(nrows=len(category_compositions), figsize=(10,25))

for i, category in enumerate(category_compositions):
    ax = axes[i]
    ax.set_title(category)
    ax.set_xlabel('Cotton')
    ax.set_ylabel('Pourcentage')
    composition_percentages = category_compositions[category] * 100
    composition_percentages.plot(kind='bar', ax=ax)
    
plt.tight_layout()
plt.show()
```
